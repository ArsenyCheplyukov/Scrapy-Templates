```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from sklearn.metrics import mean_squared_error
from matplotlib.pyplot import figure
from matplotlib.ticker import MultipleLocator, FuncFormatter
from scipy.optimize import curve_fit

figure_width = 18.  # cm
figure_height = 20. # cm
```

```python
k = 8.617333262e-5
def find_probability(x, fermi_energy, T):
    return 1/(np.exp((x-fermi_energy)/(k*T)) + 1)

fermi_energy = 10
temperature = 273

x_array = np.linspace(7.5, 12.5, 1000)
#($10^6$)
plt.grid(color='black', linestyle='--', linewidth=0.1, which='both')
plt.ylabel(r"$P_i$, вероятность состояния", fontsize=9, style='italic') #\dfrac{1}{a₁}$, $см^{-1}
plt.xlabel(r"Энергия, E, эВ", fontsize=9, style='italic')


for temp, name, color in [(0, "0K", "gainsboro"), (273, "273K", "lightgrey"), (5000, "5000K", "grey"), (10000, "10000K", "dimgrey"), (50000, "50000K", "black")]:
    plt.plot(x_array, find_probability(x_array, fermi_energy, temp), label=name, color=color)

#plt.axhline(y=1.4445, color='black', linestyle='--')
#plt.axvline(x=79, color='black', linestyle='--')
#print((f(10)-f(0))/10, f(0))
plt.legend()
plt.gray()
plt.title("Зависимость вероятности занимания фермиона данного\n состояния при термодинамическом равновессии \n от текущей энергии в эВ", fontsize=12, fontweight='bold', style='italic')
plt.show()
```
